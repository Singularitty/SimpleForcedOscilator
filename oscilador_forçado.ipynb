##  Oscilador Forçado sem amortecimento
- Equação do Oscilador forçado  

$\ddot{x} + \omega_{0}^{2}x=f\cos{\nu t}$

- Tem a seguinte solução:  

$x(t) = \frac{f}{\omega_{0}^2 - \nu^{2}}\sin{\frac{\nu t + \omega_{0} t}{2}}\sin{\frac{\omega_{0} t - \nu t}{2}}$  

Vamos modelar esta função com código para pudermos realizar um plot

```python
# Importar bibliotecas
import matplotlib.pyplot as plt
import numpy as np
from math import sin

# Definir funcao
# Podem alterar os valores default para verem o que acontece
def forced_oscilator(t, frequency = 1., base_force = 1., base_frequency = 1.1):
    first_term = base_force / (base_frequency**2 - frequency**2)
    return first_term * sin( (frequency*t + base_frequency*t) / 2) * sin( (base_frequency*t - frequency*t) / 2)
```

```python
# Definir valores para plot
time = np.arange(1,130,0.5)
position = np.vectorize(forced_oscilator)

# Realizar Plot
plt.title("Oscilador Forçado sem Atrito")
plt.xlabel("Tempo t")
plt.ylabel("Posição x")
plt.plot(time,position(time))
plt.show()
```

## Ressonância  
Este fenómeno ocorre quando a frenquência do sistema $\nu$ aproxima-se da frequência própria $\omega_0$  

Assim fazendo o limite: $\nu \to \omega_0$  

Obtem-se:  
- $x(t)=\frac{f}{2\omega_0}t\sin{\omega_0 t}$  

A amplitude do sistema vai aumentar linearmente com o tempo como veremos

```python
# Ressonancia
def forced_ressonance(t, base_force=1, base_frequency=1):
    return base_force / 2*base_frequency * t * sin(base_frequency*t)
```

```python
# Pontos
time = np.arange(1,120,0.5)
posicao = np.vectorize(forced_ressonance)

# Plot
plt.title("Ressonância")
plt.xlabel("Tempo t")
plt.ylabel("Posição x")
plt.plot(time,posicao(time))
plt.show()
```
